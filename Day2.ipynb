#**Basic** **Operations**

```python
events = spark.read.csv(
  "/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv",
  header=True,
  inferSchema=True
)

events.printSchema()
```

```python
display(
  events.select(
    "event_type",
    "product_id",
    "price"
  ).limit(10)
)
```

```python

# Get the count as an integer
high_price_count = events.filter("price > 100").count()
# Display the grouped count DataFrame
display(
  events.groupBy("event_type").count()
)


```

```python
top_brands = (
  events.filter("brand IS NOT NULL")
  .groupBy("brand")
  .count()
  .orderBy("count", ascending=False)
  .limit(5)
)
display(top_brands)
```
